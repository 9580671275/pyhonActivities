```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7]
y=[8,9,2,6,9,10]
x1=[5,4,8,12,14,11]
y1=[10,8,13,14,15,19]
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid(color="red")
plt.scatter(x,y,label='mines',marker='x',c='r')
plt.scatter(x1,y1,label='railways',marker='o',c='b')
plt.title("Dot graph")
plt.legend()
plt.show


```

```python
import matplotlib.pyplot as plt

x=[2,3,6,2,8,4]
y=[8,9,14,16,12,11]
plt.plot(x,y)
x1=[5,4,21,22,33,44]
y1=[10,8,12,13,14,15]
plt.plot(x1,y1,linewidth=6)
plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("lines between Distance and Time")
plt.show()
plt.grid(color="red")
plt.legend()
```

```python
import matplotlib.pyplot as plt
x=['rahul','raina','kohli','Dhoni','Rahane']
y=[20,80,100,80,50]
plt.bar(x,y)
plt.title("bar graph")
plt.grid(color="red")
plt.show()
plt.legend()
```

```python
import numpy as np
x2=np.arange(0,100,2)
y2=[i for i in range(len(x2))]
x1=np.arange(20,120,2)
y1=[i for i in range(len(x1))]
plt.title("By for loop graph")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.bar(x2,y2,label="water",color="r")
plt.bar(x1,y1,label="salt",color="g")
plt.legend()
plt.grid(True,color='y')
plt.show()
```
